```python
import pandas as pd

# Load the races.csv file
races_df = pd.read_csv('Archive (4)\\races.csv')

# Display the first few rows and the columns to understand its structure
races_df.head(), races_df.columns
```

```python
races_df.info()
```

```python
races_df.isnull().sum()
```

```python
races_cleaned = races_df.drop(columns=['url', 'fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 
                                       'fp3_date', 'fp3_time', 'sprint_date', 'sprint_time'])

print(races_cleaned.info())
```

```python
# Check for actual NaN values
print(races_cleaned.isnull().sum())
print(races_cleaned.isin(['NA', 'N/A', 'None', '']).sum())
```

```python
races_cleaned.rename(columns={
    'year': 'race_year',
    'round': 'race_round',
    'circuitId': 'race_circuitId',
    'name': 'race_name',
    'date': 'race_date',
    'time': 'race_time',
    'quali_date': 'race_quali_date',
    'quali_time': 'race_quali_time'
}, inplace=True)
races_cleaned.info()
```

```python
# Combine date and time columns safely, handling '\N' or missing values
races_cleaned['race_datetime'] = pd.to_datetime(
    races_cleaned['race_date'].str.strip() + ' ' + races_cleaned['race_time'].str.strip(),
    errors='coerce'
)

races_cleaned['quali_datetime'] = pd.to_datetime(
    races_cleaned['race_quali_date'].str.strip() + ' ' + races_cleaned['race_quali_time'].str.strip(),
    errors='coerce'
)
```

```python
print(races_cleaned['race_datetime'].isna().sum())
print(races_cleaned['quali_datetime'].isna().sum())
races_cleaned = races_cleaned.dropna(subset=['race_datetime', 'quali_datetime'])
   
print(races_cleaned.isnull().sum())
print(f"Data after dropping rows with missing dates: {races_cleaned.shape}")
print(races_cleaned.info())
```

```python
results_df.rename(columns={'raceId_results': 'raceId'}, inplace=True)
```

```python
overlap = set(results_df.columns).intersection(set(races_cleaned.columns))
print("Overlapping columns:", overlap)
```

```python
merged_df = results_df.merge(races_cleaned, on='raceId', how='left')
merged_df.head()
```
